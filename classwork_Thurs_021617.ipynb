```python
def makeBirdDict(formatted_nature21074-f1.csv):
    #this function opens the data file from eBird_Taxonomy_v2016_9Aug2016.csv and ext
    birdDict{}
    fobj = open(formatted_nature21074-f1.csv, "r", encoding = "ISO-8859-15") #open the file
    all_records = fobj.readlines()[1:] #read the file
    for line in all_records:
        record_elements = line.split(",")
        species = record_elements[0]
        family = record_elements{2}
        birdDict[species] = family
    #open the file
    #read the file
    #split line on comma
    #stick species and family into birdDict
    
    return birdDict
```

```python
thomas_birds = makeBirdDict(formatted_nature21074-f1.csv)
species = list(formatted_nature21074-f1.csv.keys())
#sorted (species)
species.sort()
first_species = species[0]
first_fam = thomas_birds[first_species]
print(first_species, first_fam)
print("family of " + species[0] + "is" + thomas_birds[species[0]])
```
